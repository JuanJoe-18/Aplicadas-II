#Vector gradiente

Si $f$ es una función de dos variables $x$ y $y$ cuyas derivadas parciales $f_x$ y $f_y$ existen. entonces el vector gradiente de $f$ se define como

$$\nabla f(x,y) = (f_x(x,y),f_y(x,y)).$$

Si $f$ es una función de tres variables $x$, $y$ y $z$ con derivadas parciales $f_x$, $f_y$ y $f_z$, entonces el vector gradiente de $f$ se define como

$$ \nabla f(x,y,z) = (f_x(x,y,z), f_y(x,y,z), f_z(x,y,z)). $$

**Ejemplo 1**

Calcule o encuentre el vector gradiente para la función $f(x,y)=xe^{xy}$.

**Solución**

Para encontrar el vector gradiente debemos calcular las derivadas parciales de primer orden, en este caso

$$ f_x(x,y)= (xe^{xy})_x = 1e^{xy} +xe^{xy}y = e^{xy} +xye^{xy} $$

y

$$ f_y(x,y)= (xe^{xy})_y = x(e^{xy})_y = xe^{xy}x = x^2e^{xy}. $$

Con lo anterior, tenemos

$$ \nabla f = (e^{xy} +xye^{xy}, x^2e^{xy}) $$

#Curvas de nivel

Una curva de nivel $c$ para una función $f(x,y)$ es un conjunto de puntos $(x,y) \in \mathbb{R}^2$ que satisfacen la ecuación

$$ f(x,y) = c $$

#Derivada direccional

La **derivada direccional** de una función $f(x,y)$ en el punto $(a,b)$ en la dirección del vector unitario $u=(\cos{\theta} , \sin{\theta})$, está dada o definida por

$$ D_uf(a,b) = \displaystyle\lim_{h \to 0}\dfrac{f(a+h\cos{\theta}, b+h\sin{\theta})-f(a,b)}{h} $$

siempre que este límite exista.

**Nota:**


Si $u$ es un vector, $u \neq 0$, entonces un vector unitario $v$, en la dirección y sentido del vector $u$, es

$$ v = \dfrac{u}{\|u\|}. $$

Un vector unitario $w$ en la dirección y sentido opuesto al vector $u$, es

$$ w = -\dfrac{u}{\|u\|}. $$

**Teorema**

Si $f(x,y)$ es una función diferenciable y $u$ es un vector unitario, entonces

$$ D_uf(x,y) = \nabla f(x,y)\cdot u $$

**Ejemplo 4.**

Si $f$ es la función

$$ f(x,y)=\sqrt{\dfrac{x}{y}}. $$

Calcule la derivada direccional en el punto P de coordenadas $(1,1)$ en la dirección del vector
$$v=(1,1).$$


**Solución**

Primero calculamos las derivadas parciales

$$ f_x(x,y) = \dfrac{1}{2\sqrt{x}\sqrt{y}},   \ \ \ \ f_x(1,1)=\dfrac{1}{2}$$

$$ f_y(x,y) = -\dfrac{\sqrt{x}}{2\sqrt{y^3}}, \ \ \ \ f_y(1,1)=-\dfrac{1}{2} $$

La dirección que brinda el problema no es unitaria, entonces debemos encontrar un vector que tenga la misma dirección que $v$ y sea unitario, de acuerdo a lo demostrado en clase, podemos considar el vector $u$ como

$$ u = \dfrac{v}{\|v\|} = \dfrac{(1,1)}{\sqrt{2}} = (1/\sqrt{2}, 1/\sqrt{2})$$

finalmente

$$ D_uf(1,1)=\nabla f(1,1)\cdot (1/\sqrt{2}, 1/\sqrt{2})=(1/2,-1/2)\cdot (1/\sqrt{2}, 1/\sqrt{2}) =0$$

#Valor máximo de la derivada direccional

Suponga que $f$ es una función de dos variables, el ángulo $\theta$ entre los vectores $\nabla f$ y el vector unitario $u$, está dado por la expresión

$$ \cos{\theta} = \dfrac{\nabla f \cdot u}{\|\nabla f\|\|u\|} $$

de donde

$$ \nabla f \cdot u = \|\nabla f\|\|u\|\cos{\theta} $$

de acuerdo al último teorema, tenemos que

$$ D_u f(x,y)  = \|\nabla f(x,y)\|\|u\|\cos{\theta}. $$

Ahora, si $\theta = 0$ los vectores $\nabla f$ y $u$ son paralelos y tienen la misma dirección y $\cos{\theta}$ toma su máximo valor que es $1$. En otras palabras

**El valor máximo de la derivada direccional es $\|\nabla f\|$ y ocurre cuando $u$ tiene la misma dirección que $\nabla f$ (cuando $\cos{\theta} = 1$).**


Ahora, si $\theta = \pi$ los vectores $\nabla f(x,y)$ y $u$ son paralelos y tienen dirección opuesta y $\cos{\theta}$ toma su mínimo valor que es $-1$. En otras palabras

**El valor mínimo de la derivada direccional es $-\|\nabla f(x,y)\|$ y ocurre cuando $u$ tiene la dirección opuesta a la del vector $\nabla f$ (cuando $\cos{\theta} = -1$).**


**NOTA:**

**El vector gradiente $\nabla f$ apunta en la dirección en la cual $f$ crece con mayor rapidez, en tanto que $-\nabla f$ apunta en la dirección en la cual $f$ decrece con mayor rapidez.**
